```python
df = pd.read_csv("train_data.csv")
df.head()
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df.head()
```

```python
null_sum = df.isnull().sum()
null_sum
```

```python
def null_percentage(df):
    for i,name in enumerate(df.columns):
        if null_sum[i] > 0:
            print(f"The Percentage of Null Values in {name} is {(null_sum[i]/df.shape[0])*100} %")
```

```python
null_percentage(df)
```
